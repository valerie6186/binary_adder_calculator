Step 1: Define Basic Logic Gates
</div> 

```python
def AND(a, b):
    return int(a and b)


def OR(a, b):
    return int(a or b)


def XOR(a, b):
    return int(a != b)
```

<div style="background-color: #f3e5ff; padding: 12px; border-radius: 10px; font-size: 17px; font-weight: bold; font-family: Times;">
we need to define the three logic gates needed to build the binary adder:</div>

* <div style="font-family: Times; font-size: 16px">the AND gate calculates the carry bit because it returns 1 only when both inputs are 1</div>

* <div style="font-family: Times; font-size: 16px">the OR gate combines possible carry values and determines whether a carry should be passed to the next binary column</div>

* <div style="font-family: Times; font-size: 16px">the XOR gate calculates the sum bit because it returns 1 when the two inputs are different and 0 when they are the same</div>


<div style="background-color: #ffe6f2; padding: 12px; border-radius: 10px; font-size: 16px; font-family: Times;">
<h3 style="font-size: 20px; font-weight: bold;">Step 2: Create Half Adder</h3>
<ul> <li>this adds two binary digits, producing sum and carry values </li> </ul>
</div>

```python
def half_adder(a, b):
    sum_bit = XOR(a, b)
    carry_bit = AND(a, b)

    return {
        "sum": sum_bit,
        "carry": carry_bit
    }
```

<div style="background-color: #f3e5ff; padding: 12px; border-radius: 10px; font-size: 17px; font-weight: bold; font-family: Times;">
test/verify half adder:</div>

* <div style="font-family: Times; font-size: 16px">the AND gate calculates the carry bit because it returns 1 only when both inputs are 1</div>
* <div style="font-family: Times; font-size: 16px">the XOR gate calculates the sum bit because it returns 1 when the two inputs are different and 0 when they are the same</div>


```python
print(half_adder(0, 0))
print(half_adder(0, 1))
print(half_adder(1, 0))
print(half_adder(1, 1))
```

<div style="background-color: #ffe6f2; padding: 12px; border-radius: 10px; font-size: 16px; font-family: Times;">
<h3 style="font-size: 20px; font-weight: bold;">Step 3: Create Full Adder</h3>
<ul> <li>this takes three inputs: two binary digits and a potential carry in, producing sum and carry out values </li> </ul>
</div>

```python
def full_adder(a, b, carry_in):
    first_half = half_adder(a, b)
    second_half = half_adder(first_half["sum"], carry_in)

    carry_out = OR(
        first_half["carry"],
        second_half["carry"]
    )

    return {
        "sum": second_half["sum"],
        "carry": carry_out
    }
```

* <div style="font-family: Times; font-size: 16px">the OR gate combines possible carry values and determines whether a carry should be passed to the next binary column</div>


```python
print(full_adder(0, 0, 0))
print(full_adder(0, 1, 0))
print(full_adder(1, 1, 0))
print(full_adder(1, 1, 1))
```

Step 4: Create Binary Addition Function</div> 

```python
def binary_add(binary_a, binary_b):
    if not binary_a or not binary_b:
        raise ValueError("both binary numbers are required")

    if not set(binary_a).issubset({"0", "1"}):
        raise ValueError("the first number must contain only 0 and 1")

    if not set(binary_b).issubset({"0", "1"}):
        raise ValueError("the second number must contain only 0 and 1")

    length = max(len(binary_a), len(binary_b))

    binary_a = binary_a.zfill(length)
    binary_b = binary_b.zfill(length)

    result = []
    steps = []
    carry = 0

    # line starts at right-most character and moves backward
    for position in range(length - 1, -1, -1):
        # current digits are converted into integers
        a_bit = int(binary_a[position])
        b_bit = int(binary_b[position])
        carry_in = carry

        # full adder adds the digits and carry
        addition = full_adder(a_bit, b_bit, carry)

        # store the sum bit
        result.append(str(addition["sum"]))
        # new carry is passed into the next column 
        carry = addition["carry"]
        # shows the steps of the addition
        steps.append({
            "position": position,
            "a": a_bit,
            "b": b_bit,
            "carry_in": carry_in,
            "sum": addition["sum"],
            "carry_out": carry
        })

    if carry == 1:
        result.append("1")

    # since the result was calculated backwards, reversing fixes the format
    result.reverse()

    return {
        "first_number": binary_a,
        "second_number": binary_b,
        "binary_sum": "".join(result),
        "final_carry": carry,
        "steps": steps
    }
```

```python
answer = binary_add("1011", "0110")

print("first number:", answer["first_number"])
print("second number:", answer["second_number"])
print("binary sum:", answer["binary_sum"])
```

Step 5: Create Binary Adder Calculator Interface</div> 

```python
import ipywidgets as widgets
from IPython.display import display
```

<div style="background-color: #e6fff2; padding: 12px; border-radius: 10px; font-size: 16px; font-family: Times;">
<ul> <li>'ipywidgets' provides interactive tools such as text boxes and buttons</li> </ul>
<ul> <li>'display' allows the completed calculator interface to appear inside Jupyter Notebook</li> </ul>
</div>

```python
first_input = widgets.Text(
    value="",
    placeholder="enter the first binary number",
    description="first:"
)

second_input = widgets.Text(
    value="",
    placeholder="enter the second binary number",
    description="second:"
)
```

<div style="background-color: #e6fff2; padding: 12px; border-radius: 10px; font-size: 16px; font-family: Times;">
<ul> <li>creates a text box for the first and second binary numbers</li> </ul>
</div>

```python
add_button = widgets.Button(
    description="Add Binary Numbers",
    button_style="success"
)
```

<div style="background-color: #e6fff2; padding: 12px; border-radius: 10px; font-size: 16px; font-family: Times;">
<ul> <li>creates button for adding the numbers together</li> </ul>
<ul> <li>'description' controls the words displayed on the button</li> </ul>
<ul> <li>'button_style="success"' makes the button green</li> </ul>
</div>

```python
output = widgets.Output()
```

<div style="background-color: #e6fff2; padding: 12px; border-radius: 10px; font-size: 16px; font-family: Times;">
<ul> <li>creates area where the calculator will display: the binary numbers, final sum, addition steps, error messages</li> </ul>
</div>

<div style="background-color: #ffe6f2; padding: 12px; border-radius: 10px; font-size: 20px; font-weight: bold; font-family: Times;">
Step 6: Program Addition Button</div> 

```python
def calculate_binary_sum(button):
    # means everything printed inside this section will appear in the calculator’s output box instead of underneath
    with output:
        
        # clears old result before showing new one
        output.clear_output() 

        # reads values & removes accidental spaces before or after the number
        first_number = first_input.value.strip()
        second_number = second_input.value.strip()

        # runs code that may produce an error
        try:
            # calls binary addition function, dictionary containing information is stored here
            answer = binary_add(
                first_number,
                second_number
            )

            # prints main calculation
            print("binary addition")
            print()
            print("  ", answer["first_number"])
            print("+ ", answer["second_number"])
            
            # creates the dividing line
            print("-" * (len(answer["binary_sum"]) + 2))
            print("  ", answer["binary_sum"])
            print()

            print("addition steps")

            # prints a heading before the individual column calculations
            for step in answer["steps"]:
                print(
                    step["a"],
                    "+",
                    step["b"],
                    "+ carry",
                    step["carry_in"],
                    "=",
                    step["sum"],
                    "carry",
                    step["carry_out"]
                )

        # handles invalid inputs
        except ValueError as error:
            print("error:", error)

# connect function to the button
add_button.on_click(calculate_binary_sum)
```
